```python
df = df.drop(columns=["customerID"])
```

```python
df.dropna(inplace=True)
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
cat_vars = ["PhoneService", "Contract", "PaperlessBilling","PaymentMethod"]

num_vars = ["tenure", "MontlyCharges", "TotalCharges"]

modeling_vars = cat_vars + num_vars
```

```python
x = df.filter(modeling_vars).copy()

y = df["Churn"].copy()

y = pd.DataFrame(y)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(y.Churn)
y.Churn = le.transform(y.Churn)
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()



le.fit(y.Churn)

y.Churn = le.transform(y.Churn)

```

```python
x = pd.get_dummies(x)
```

```python
contagem_churn = df["Churn"].value_counts()
print(contagem_churn)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(X_train_scaled,columns = X_train.columns)
X_test = pd.DataFrame(X_test_scaled,columns = X_test.columns)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

clf = MLPClassifier( hidden_layer_sizes=(100, 50, 20),activation='logistic')
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred) * 100

f1 = f1_score(y_test, y_pred)

print(f"Acurácia: {accuracy:.2f}%")
print(f"Escore F1: {f1:.2f}")
```
